```python
#to check the target variable ---unique() is used
data['diagnosis'].unique()
```

```python
#converting the object values to numerical values in the target variable
classes={'M':0,'B':1}
data.replace({'diagnosis':classes},inplace=True)
print(data)
```

```python
#x contains all the features 
x=data.drop(['diagnosis'],axis=1)
x.head(5)
```

```python
#y contains the target variable
y=data['diagnosis']
y.head(5)
```

```python
#split the universal data
#library:sklearn
#module:model_selection
#class:train_test_split
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y)
```

```python
#algorithm selection
#library:sklearn
#module:linear_model
#class:logistic regression
from sklearn.linear_model import LogisticRegression as logreg
model_logreg=logreg()
```

```python
#Train the model with training set of data
#Function: fit(arrays of x and y train)
model_logreg.fit(x_train,y_train)
```

```python
#TEST THE MODEL
#1. Ask the model to predict answers for test data
#2. Check accuracy for test data

#PREDICT
#Function: predict()
y_pred=model_logreg.predict(x_test)
```

```python
#ACCURACY
#Function: score()
logreg_acc=model_logreg.score(x_test, y_test)
print(logreg_acc)
```

```python
#CONFUSION MATRIX
#Ratio of right and wrong predictions
#Library: sklearn
#Module : metrics
#Class  : confusion_matrix
from sklearn.metrics import confusion_matrix as cm
conmat=cm(y_test,y_pred)
print(conmat)
```
